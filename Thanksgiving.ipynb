```python
import pandas as pd
from afinn import Afinn
```

```python
df = pd.read_csv('Thanksgiving.csv', sep=",")
df
```

```python
afinn = Afinn()
sentiment_info = {}
for index, row in df.iterrows():
    if row['date'] not in sentiment_info:
        sentiment_info[row['date']] = [int(afinn.score(str(row['text']))),1]
    else:
        sentiment_info[row['date']][0] += int(afinn.score(str(row['text'])))
        sentiment_info[row['date']][1] += 1
```

```python
print(sentiment_info)
#[sum of sentiment, No. of tweets]
```

```python
sentiment_data = {}
for date in sentiment_info:
    sentiment_data[date] = (sentiment_info[date][0]/float(sentiment_info[date][1]))
```
